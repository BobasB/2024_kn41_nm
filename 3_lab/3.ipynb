### Статистичні операції
- статистичні дані нам потрібні як приклад знаходження екстремумів (мінімумів та максимумів)

```python
import numpy as np

ss = np.random.randint(0, 50, size=(4, 4))

print(ss, 
      "\nМінімум мав би бути 0, а є:", 
      ss.min(),
      "\nМаксимум мав би бути 49, а маємо: ",
      ss.max())

print(ss.shape, 
      "\nБеремо мініальне значення по кожному стовпцю:", 
      ss.min(axis=0),
      "\nБеремо мін поткожному рядку", 
      ss.min(axis=1))
```

### Чат зробив на приклад

```python
import numpy as np
import matplotlib.pyplot as plt

# Генеруємо значення x від 0 до 4π з кроком 0.01
x = np.arange(0, 4 * np.pi, 0.01)
# Обираємо функцію: синус
y = np.sin(x)**2 * np.cos(x-0.5*np.pi)

# Знаходимо максимальне та мінімальне значення функції
y_max = y.max()
y_min = y.min()

# Знаходимо відповідні значення x для мінімуму та максимуму
x_max = x[y.argmax()]
x_min = x[y.argmin()]
print(x_max, y_max)

# Створюємо графік
plt.figure(figsize=(10, 5))
plt.plot(x, y, label="sin(x)", color="blue")

# Додаємо точки максимуму і мінімуму та прямі лінії
plt.plot(x_max, y_max, 'ro', label="Max")
plt.plot(x_min, y_min, 'go', label="Min")
plt.axhline(y=y_max, color='red', linestyle='dotted', linewidth=0.8)
plt.axhline(y=y_min, color='green', linestyle="--", linewidth=0.8)
plt.axvline(x=x_max, color='red', linestyle='dashdot', linewidth=0.8)
plt.axvline(x=x_min, color='green', linestyle="--", linewidth=0.8)

# Налаштовуємо підписи до осей та заголовок графіка
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("Графік функції sin(x) з максимальними та мінімальними точками")
plt.legend()

# Відображаємо графік
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Генерація зарплат із невеликою варіацією
np.random.seed(42)  # Для повторюваності результатів
junior_salaries = np.random.normal(loc=500, scale=50, size=100)  # Джуни
middle_salaries = np.random.normal(loc=1000, scale=100, size=50)  # Мідли
senior_salaries = np.random.normal(loc=5000, scale=500, size=10)  # Сеньйори

# Об'єднуємо всі зарплати в один масив
salaries = np.concatenate([junior_salaries, middle_salaries, senior_salaries])

# Корекція зарплат, щоб вони не були від’ємними
salaries = np.maximum(salaries, 0)

# Розрахунок середнього, медіанного значень, дисперсії та стандартного відхилення
mean_salary = np.mean(salaries)
median_salary = np.median(salaries)
variance_salary = np.var(salaries)
std_salary = np.std(salaries)

# Вивід значень
print(f"Середня зарплата (Mean): {mean_salary:.2f}")
print(f"Медіана зарплати (Median): {median_salary:.2f}")
print(f"Дисперсія (Variance): {variance_salary:.2f}")
print(f"Стандартне відхилення (Std): {std_salary:.2f}")

# Візуалізація
plt.figure(figsize=(10, 6))

# Гістограма зарплат
plt.hist(salaries, bins=30, alpha=0.7, color='skyblue', edgecolor='black', label='Розподіл зарплат')

# Додаткові лінії на графіку
plt.axvline(mean_salary, color='green', linestyle='--', label=f'Середнє (Mean): {mean_salary:.2f}')
plt.axvline(median_salary, color='blue', linestyle='--', label=f'Медіана (Median): {median_salary:.2f}')


plt.axvline(max(mean_salary - std_salary, np.min(salaries)), color='red', linestyle=':', label=f'1 Std нижче: {max(mean_salary - std_salary, 0):.2f}')

plt.axvline(mean_salary + std_salary, color='red', linestyle=':', label=f'1 Std вище: {mean_salary + std_salary:.2f}')
plt.axvline(np.min(salaries), color='purple', linestyle='-', label=f'Мінімум: {np.min(salaries):.2f}')
plt.axvline(np.max(salaries), color='orange', linestyle='-', label=f'Максимум: {np.max(salaries):.2f}')

# Додаткові параметри графіку
plt.title('Розподіл зарплат IT-спеціалістів')
plt.xlabel('Зарплата ($)')
plt.ylabel('Кількість')
plt.legend()
plt.grid(alpha=0.3)

# Показуємо графік
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Параметри для моделювання гравців
#np.random.seed(42)  # Для повторюваності
num_players = 1000  # Кількість гравців
max_level = 100     # Максимальний рівень

# Моделюємо рівні гравців із експоненційним розподілом
levels = np.random.exponential(scale=max_level / 10, size=num_players).astype(int)
levels = np.clip(levels, 1, max_level)  # Корекція, щоб рівні були від 1 до max_level

# Визначаємо топ гравців (гравці з рівнем 90+)
top_players = levels[levels >= 90]

# Розрахунок статистичних характеристик
mean_level = np.mean(levels)
median_level = np.median(levels)
variance_level = np.var(levels)
std_level = np.std(levels)

# Вивід статистичних даних
print(f"Середній рівень (Mean): {mean_level:.2f}")
print(f"Медіанний рівень (Median): {median_level:.2f}")
print(f"Дисперсія рівнів (Variance): {variance_level:.2f}")
print(f"Стандартне відхилення рівнів (Std): {std_level:.2f}")
print(f"Топ гравців (рівень 90+): {len(top_players)}")

# Візуалізація рівнів гравців
plt.figure(figsize=(12, 8))

# Гістограма всіх рівнів
plt.hist(levels, bins=30, alpha=0.7, color='skyblue', edgecolor='black', label='Рівні всіх гравців')

# Додаткові вертикальні лінії
plt.axvline(mean_level, color='green', linestyle='--', label=f'Середній рівень (Mean): {mean_level:.2f}')
plt.axvline(median_level, color='blue', linestyle='--', label=f'Медіанний рівень (Median): {median_level:.2f}')
plt.axvline(np.min(levels), color='purple', linestyle='-', label=f'Мінімальний рівень: {np.min(levels):.2f}')
plt.axvline(np.max(levels), color='orange', linestyle='-', label=f'Максимальний рівень: {np.max(levels):.2f}')
plt.axvline(mean_level - std_level, color='red', linestyle=':', label=f'1 Std нижче: {max(mean_level - std_level, 0):.2f}')
plt.axvline(mean_level + std_level, color='red', linestyle=':', label=f'1 Std вище: {mean_level + std_level:.2f}')

# Графік рівнів топ гравців
plt.scatter(np.arange(len(top_players)), top_players, color='darkblue', label='Топ гравці (рівень 90+)', alpha=0.8)

# Підписуємо графік
plt.title('Рівні гравців у грі Lineage', fontsize=16)
plt.xlabel('Рівень', fontsize=12)
plt.ylabel('Кількість гравців', fontsize=12)
plt.legend(fontsize=10)
plt.grid(alpha=0.3)

# Показуємо графік
plt.show()
```
