### Статистичні операції
- статистичні дані нам потрібні як приклад знаходження екстремумів (мінімумів та максимумів)

```python
import numpy as np

ss = np.random.randint(0, 50, size=(4, 4))

print(ss, 
      "\nМінімум мав би бути 0, а є:", 
      ss.min(),
      "\nМаксимум мав би бути 49, а маємо: ",
      ss.max())

print(ss.shape, 
      "\nБеремо мініальне значення по кожному стовпцю:", 
      ss.min(axis=0),
      "\nБеремо мін поткожному рядку", 
      ss.min(axis=1))
```

### Чат зробив на приклад

```python
import numpy as np
import matplotlib.pyplot as plt

# Генеруємо значення x від 0 до 4π з кроком 0.01
x = np.arange(0, 4 * np.pi, 0.01)
# Обираємо функцію: синус
y = np.sin(x)**2 * np.cos(x-0.5*np.pi)

# Знаходимо максимальне та мінімальне значення функції
y_max = y.max()
y_min = y.min()

# Знаходимо відповідні значення x для мінімуму та максимуму
x_max = x[y.argmax()]
x_min = x[y.argmin()]
print(x_max, y_max)

# Створюємо графік
plt.figure(figsize=(10, 5))
plt.plot(x, y, label="sin(x)", color="blue")

# Додаємо точки максимуму і мінімуму та прямі лінії
plt.plot(x_max, y_max, 'ro', label="Max")
plt.plot(x_min, y_min, 'go', label="Min")
plt.axhline(y=y_max, color='red', linestyle='dotted', linewidth=0.8)
plt.axhline(y=y_min, color='green', linestyle="--", linewidth=0.8)
plt.axvline(x=x_max, color='red', linestyle='dashdot', linewidth=0.8)
plt.axvline(x=x_min, color='green', linestyle="--", linewidth=0.8)

# Налаштовуємо підписи до осей та заголовок графіка
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("Графік функції sin(x) з максимальними та мінімальними точками")
plt.legend()

# Відображаємо графік
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Генерація зарплат із невеликою варіацією
np.random.seed(42)  # Для повторюваності результатів
junior_salaries = np.random.normal(loc=500, scale=50, size=100)  # Джуни
middle_salaries = np.random.normal(loc=1000, scale=100, size=50)  # Мідли
senior_salaries = np.random.normal(loc=5000, scale=500, size=10)  # Сеньйори

# Об'єднуємо всі зарплати в один масив
salaries = np.concatenate([junior_salaries, middle_salaries, senior_salaries])

# Розрахунок середнього та медіанного значення
mean_salary = np.mean(salaries)
median_salary = np.median(salaries)

print(f"Середня зарплата: {mean_salary} а Медіана: {mean_salary}")

# Візуалізація
plt.figure(figsize=(12, 6))

# Гістограма зарплат
plt.hist(salaries, bins=30, color='skyblue', edgecolor='black', alpha=0.7, label="Розподіл зарплат")

# Додаємо вертикальні лінії для середнього та медіанного значення
plt.axvline(mean_salary, color='red', linestyle='--', linewidth=1.5, label=f"Середнє значення: {mean_salary:.2f}")
plt.axvline(median_salary, color='green', linestyle='--', linewidth=1.5, label=f"Медіанне значення: {median_salary:.2f}")

# Налаштування графіка
plt.title("Розподіл зарплат IT-спеціалістів", fontsize=16)
plt.xlabel("Зарплата ($)", fontsize=12)
plt.ylabel("Кількість спеціалістів", fontsize=12)
plt.legend(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Відображаємо графік
plt.show()
```
